```python
# 最小年月和最大年月
start_date = '2020-08'
end_date = '2023-04'

# 生成日期範圍並轉換為所需的字符串格式
date_range = pd.date_range(start=start_date, end=end_date, freq='M').strftime('%Y-%m').tolist()

# 創建字典，將每個日期設置為0
result_dict = {date: 0 for date in date_range}
result_dict
```

```python
grouped = pd.DataFrame(df.groupby(df['p_year_month'])['薪資'].sum()).reset_index()
grouped
for index, row in grouped.iterrows():
    result_dict[row['p_year_month']] = row['薪資']
    # print(row['p_year_month']) # 输出每行的索引值
# for k,v in grouped['薪資']
result_dict
```
